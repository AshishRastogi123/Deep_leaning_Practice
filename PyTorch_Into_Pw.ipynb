# Tensor Operations and Gradients

```python
#Create tensor
x=torch.tensor(3.)
w=torch.tensor(4. , requires_grad=True) # require_grad=True mean we tell py torch to automatic compute derivative of w in future for
#It's an Pytorch feature
b=torch.tensor(5., requires_grad=True)
x,w,b
```

```python
# Airthemetic Operations
y=w*x + b
y
```

```python
#Compute Derivatives
y.backward()
```

```python
#display The value
print("dy/dx : ",x.grad)
print("dy/dw : ", w.grad)
print("dy/db : ", b.grad)
```
